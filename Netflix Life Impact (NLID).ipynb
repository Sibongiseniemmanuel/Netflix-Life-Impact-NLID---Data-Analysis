```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv(r"C:\Users\nteny\Desktop\DATASETS\Netflix Life Impact Dataset (NLID).csv", delimiter=';')

df.head()
```

```python

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
df
```

```python
df["Movie Title"]=df["Movie Title"].replace("Ma Raineys Black Bottom","Ma Rainey's Black Bottom")
df.head()
```

```python
df[["Suggested Percentage","Suggested"]]=df["Suggested to Friends/Family (Y/N %)"]\
    .str.extract(r"(\d+)%\s*(Y|N)",expand=True)
```

```python
df.drop(columns=["Suggested to Friends/Family (Y/N %)"],inplace=True)
df.head()
```

```python

df['Genre'] = df['Genre'].apply(lambda x: x.replace("['", "").replace("']", "").replace("\\", "").replace("'", '').strip())
df['Genre'] = df['Genre'].str.split('/')
df1 = df.explode('Genre')

df1['Genre'] = df1['Genre'].str.strip()
genre_reviews = df1.groupby('Genre')['Number of Reviews'].sum().sort_values(ascending=False)
print(genre_reviews.head())  
```

```python

df1 = df1.reset_index(drop=True)

df1.head()
```

```python
df1['Suggested Percentage'] = df1['Suggested Percentage'].str.replace('%', '').astype(float)
```

```python
df1.to_csv(r"C:\Users\nteny\Desktop\DATASETS\Netflix_Life_Impact_Dataset_v2.csv")
```

```python

top10_suggested_ = df1.nlargest(10, 'Suggested Percentage')

# Plot
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(18, 6))
sns.barplot(x='Suggested Percentage', y='Movie Title', data=top10_suggested_, palette='viridis')
plt.title('Top 10 Most Suggested Netflix Movies')
plt.xlabel('Suggested Percentage')
plt.ylabel('Movie Title')
plt.tight_layout()

# Save the plot image
plt.savefig(r"C:\Users\nteny\Desktop\DATASETS\my_SUGGESTED_plot.png")
plt.show()
```

```python
# Get top 10 movies by average rating
top10_suggested = df1.nlargest(10, 'Average Rating')

# Create the plot
plt.figure(figsize=(18, 6))  
sns.lineplot(x='Average Rating', y='Movie Title', data=top10_suggested, marker='o')  

# Display the plot
plt.title("Top 10 Netflix Movies that a hightly rated")
plt.xlabel("Average Rating")
plt.ylabel("Movie Title")
plt.tight_layout()
plt.savefig(r"C:\Users\nteny\Desktop\DATASETS\my_RATED_plot.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by Genre and sum the number of reviews
genre_reviews = df1.groupby('Genre')['Number of Reviews'].sum().sort_values(ascending=False).reset_index()

# Plot
plt.figure(figsize=(14, 8))
sns.barplot(x='Number of Reviews', y='Genre', data=genre_reviews, palette='viridis')
plt.title('Most Watched Genres by Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Genre')
plt.tight_layout()
plt.savefig(r"C:\Users\nteny\Desktop\DATASETS\my_GENRE_plot.png")

plt.show()
```
